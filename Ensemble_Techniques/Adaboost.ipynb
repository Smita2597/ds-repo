#### Import Necessary Library

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
cancer_data = datasets.load_breast_cancer()
```

```python
cancer_data_df = pd.DataFrame(data=cancer_data.data,columns=cancer_data.feature_names)
```

```python
cancer_data_df['target'] = cancer_data['target']
```

#### Data Understanding

#### Model Building

```python
x = cancer_data_df.drop('target',axis=1)
y = cancer_data_df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=12,stratify=y)
```

```python
y_train.value_counts()
```

```python
rf_model = RandomForestClassifier()
ab_model = AdaBoostClassifier(learning_rate=1)
gb_model = GradientBoostingClassifier()
```

```python
rf_model.fit(x_train,y_train)
ab_model.fit(x_train,y_train)
gb_model.fit(x_train,y_train)
```

### Model Testing|Model Evaluation

```python
rf_model_y_pred_train = rf_model.predict(x_train)
rf_model_y_pred_test = rf_model.predict(x_test)
```

```python
ab_model_y_pred_train = ab_model.predict(x_train)
ab_model_y_pred_test = ab_model.predict(x_test)
```

```python
gb_model_y_pred_train = gb_model.predict(x_train)
gb_model_y_pred_test = gb_model.predict(x_test)
```

```python
print("RF Model Evaluation:")
print("Training Accuracy:",accuracy_score(y_train,rf_model_y_pred_train))
print("Testing Accuracy:",accuracy_score(y_test,rf_model_y_pred_test))
print("precision_score:",precision_score(y_test,rf_model_y_pred_test))
print("Recall_score:",recall_score(y_test,rf_model_y_pred_test))
print("Confusion Metrics:",confusion_matrix(y_test,rf_model_y_pred_test))
```

```python
print("Adaboost Model Evaluation:")
print("Training Accuracy:",accuracy_score(y_train,ab_model_y_pred_train))
print("Testing Accuracy:",accuracy_score(y_test,ab_model_y_pred_test))
print("precision_score:",precision_score(y_test,ab_model_y_pred_test))
print("Recall_score:",recall_score(y_test,ab_model_y_pred_test))
print("Confusion Metrics:",confusion_matrix(y_test,ab_model_y_pred_test))
```

```python
print("Gradient Boosting Model Evaluation:")
print("Training Accuracy:",accuracy_score(y_train,gb_model_y_pred_train))
print("Testing Accuracy:",accuracy_score(y_test,gb_model_y_pred_test))
print("precision_score:",precision_score(y_test,gb_model_y_pred_test))
print("Recall_score:",recall_score(y_test,gb_model_y_pred_test))
print("Confusion Metrics:",confusion_matrix(y_test,gb_model_y_pred_test))
```
