```python
import pandas as pd
from sklearn.feature_selection import RFE,SelectFromModel,chi2,SelectKBest
from sklearn.linear_model import LogisticRegression
```

### Read Data

```python
diabetes_df = pd.read_csv('pima-indians-diabetes.data.csv')
```

```python
filename = "pima-indians-diabetes.data.csv"
feature_name = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

```python
datafarme = pd.read_csv(filename)
array = datafarme.values
array
```

```python
diabetes_df = pd.DataFrame(data = array,columns=feature_name)
```

#### Feature Selection using SelectKBest

```python
x = diabetes_df.drop('class',axis =1)
y = diabetes_df['class']
```

```python
model = SelectKBest(score_func=chi2,k=4)
fited_model = model.fit(x,y)
```

```python
fited_model.scores_
```

```python
(fited_model.transform(x))
```

#### Feature selection using RFE(Recursive Forword Elimination)

```python
log_model = LogisticRegression(max_iter=400)
```

```python
RFE_model = RFE(estimator=log_model,n_features_to_select=3)
```

```python
fit_rfe = RFE_model.fit(x,y)
```

```python
fit_rfe.n_features_
```

```python
fit_rfe.ranking_
```

```python
pd.DataFrame({'columns':x.columns,'impotance':fit_rfe.ranking_})
```
