
## Qualitative & Quantitative Type of Data

- Qualitative: We try to guage data based on its quality (e.g., taste of food, quality of a photograph)

- Quantitative: We try to guage data based on numbers/statistics (e.g., no. of cars, no. of pencils)

    - Descrete - Is is not measurable (No units)/it is just a count

    - Continuous - it is measurable (It has unit)
