### Grouping vs Pivot Table vs Crosstab

```python
insurance = pd.read_csv("insurance.csv")
```

```python
insurance.head()
```

```python
insurance.isna().sum()
```

```python
insurance.describe(include='all')
```

#### What is the average insurance charges by region

```python
insurance['region'].unique()
```

```python
insurance['region'].nunique()
```

```python
round(insurance.groupby(by='region')['charges'].mean().sort_values(ascending=False) ,2)
```

#### What is the average insurance charges by region and gender sorted in descending order?

```python
round(insurance.groupby(['region','sex'])['charges'].mean().sort_values(ascending=False),2)
```

### Pivot Table

```python
pd.pivot_table(data=insurance,values=['charges'],index=['region'],columns=['sex','smoker'])
```

```python
round(pd.pivot_table(data=insurance,values=['children','charges'],index=['region'],columns=['sex'],aggfunc='min'),0)
```

```python
round(pd.pivot_table(data=insurance,values=['children','charges'],index=['region'],columns=['sex'],aggfunc='max'),0)
```

#### Crosstab

```python
pd.crosstab(index = insurance['smoker'],columns=insurance['sex'],margins=True)
```

```python
pd.crosstab(index = insurance['smoker'],columns=insurance['sex'],margins=False)
```

### Count of children by region

```python
pd.crosstab(index=insurance['region'],columns=insurance['children'],margins=False)
```
