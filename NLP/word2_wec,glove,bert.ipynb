```python
from gensim.models import Word2Vec
```

```python
tokenised_sentence = [['Hello', 'Welcome', 'to','the', 'Data', 'Science', 'Program'],
['Hello','Glad','to','have','you','here'],
 ["Hello", 'Welcome','Back']]
```

```python
model = Word2Vec(sentences=tokenised_sentence,min_count=1)
```

```python
print(model)
```

```python
list(model.wv.index_to_key)
```

```python
model.wv.key_to_index
```

```python
model.wv.get_vector('Data')
```

```python
model.wv.get_vector('Hello')
```

```python
model.wv.most_similar('Hello')
```

### BERT - Glove

```python
from flair.embeddings import WordEmbeddings,TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```python
glove_embeddings = WordEmbeddings('glove')
```

```python
sentence = Sentence('The founder apple is Steve Jobs')
```

```python
sent_1 = glove_embeddings.embed(sentence)
```

```python
for token in sentence:
    print(token)
    print(token.embedding)
    print("...........................................................")
```

```python
sentence_2 = Sentence("Apple is a fruit")
sent_2 = glove_embeddings.embed(sentence_2)
```

```python
for token in sentence_2:
    print(token)
    print(token.embedding)
    print("...........................................................")
```

```python
bert_embeddings = TransformerWordEmbeddings('bert-base-multilingual-cased')
```

```python
bert_embeddings.embed(sentence)
```

```python
for token in sentence:
    print(token)
    print(token.embedding)
    print("...........................................................")
```
