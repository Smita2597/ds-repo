```python
### import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler,LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV
```

### Import & Read Data

```python
mushroom_df = pd.read_csv("mushroom.csv")
mushroom_df.shape
```

##### Data Visualization

```python
poisonous_df = mushroom_df[mushroom_df['class']=='poisonous']
edible_df = mushroom_df[mushroom_df['class']=='edible']
```

```python
## plot the graph to check the distribution of the classes
ax_1 = poisonous_df.plot(kind='scatter',x ='stalk_height',y='cap_diameter',color = 'red',label = 'poisonous')
edible_df.plot(kind='scatter',x ='stalk_height',y='cap_diameter',color = 'blue',label = 'edible',ax=ax_1)
plt.show()
```

### Data Preprocessing

```python
# ## Encode categorical variable
encoder  = LabelEncoder()

for i in mushroom_df.columns:
    mushroom_df[i] = encoder.fit_transform(mushroom_df[i])
```

```python
mushroom_df.head(5)
```

```python
mushroom_df.info()
```

### Model Building

```python
## Divide dataset into dependenent and independenent feature
x = mushroom_df.drop('class',axis=1)
y = mushroom_df['class']
```

```python
# ## split the dataset into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
### Apply SVM
svm = SVC()
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

#### Grid SeArch CV

```python
## Hyperparameter Tuning using GridSearch CV
param_gird = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear']}
```

```python
grid = GridSearchCV(svm,param_gird,refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = svm.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_grid))
```
