#### Corearation matrix for recommendation

```python
movie_df['userId'].unique()
```

```python
user_recom_corr = pd.pivot_table(movie_df,values='rating',index='movie',columns='userId').fillna(0)
user_recom_corr.head()
```

```python
user_recom_corr.columns = movie_df['userId'].unique()
```

```python
user_recom_corr
```

#### Euclidean distance matrix for recommendations

```python
from sklearn.metrics import pairwise_distances
```

```python
user_recom_traspose = user_recom_corr.T
```

```python
user_val = user_recom_traspose.values
```

```python
eucliden_distance_df = pd.DataFrame(pairwise_distances(X= user_val,metric='euclidean'),columns = movie_df['userId'].unique())
```

```python
eucliden_distance_df
```

```python
eucliden_distance_df
```

```python
eucliden_distance_df.iloc[:15,:15].idxmin()
```

#### Cosine Similarity for recommendation

```python
cosine_silmilar_df = 1 - pairwise_distances(X= user_val,metric='cosine')
```

```python
cosine_silmilar_df_new = pd.DataFrame(cosine_silmilar_df,columns = movie_df['userId'].unique())
```

```python
cosine_silmilar_df_new.iloc[:15,:15].idxmin()
```
