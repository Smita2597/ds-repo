### Import required lib

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
retail_df = pd.read_excel("Online Retail.xlsx")
```

### Data cleaning

```python
retail_df.drop('CustomerID',axis=1)
```

```python
retail_df.dropna(axis=0,inplace=True)
```

```python
retail_df.isnull().sum()
```

```python
retail_df['Country'].value_counts()
```

```python
# United Kingdom ,Germany, France 

uk_data = retail_df[retail_df['Country']=='United Kingdom']
Germany_data = retail_df[retail_df['Country']=='Germany']
France_data = retail_df[retail_df['Country']=='France']
```

```python
uk_data.head()
```

```python
uk_pivot =pd.pivot_table(data =uk_data,values = 'Quantity',index='InvoiceNo',columns='Description',fill_value=0)
uk_pivot = uk_pivot.applymap(lambda x: 1 if x>0.5 else 0)
```

```python
germancy_pivot =pd.pivot_table(data =Germany_data,values ="Quantity",index='InvoiceNo',columns='Description',fill_value=0)
germancy_pivot = germancy_pivot.applymap(lambda x: 1 if x>0.5 else 0)
```

```python
France_pivot =pd.pivot_table(data =France_data,values ="Quantity",index='InvoiceNo',columns='Description',fill_value=0)
France_pivot = France_pivot.applymap(lambda x: 1 if x>0.5 else 0)
```

```python
uk_support = apriori(df =uk_pivot, min_support=0.05, use_colnames=True)
uk_support
```

```python
France_support = apriori(df =France_pivot, min_support=0.05, use_colnames=True)
```

```python
Germany_support = apriori(df =germancy_pivot, min_support=0.05, use_colnames=True)
```

```python
uk_data['Description'].nunique()
```

```python
association_rules(uk_support,min_threshold = 0.01,metric='confidence',num_itemsets=3860)
```

```python
France_data['Description'].nunique()
```

```python
association_rules(France_support,min_threshold = 0.01,metric='confidence',num_itemsets=1545)
```

```python
Germany_data['Description'].nunique()
```

```python
association_rules(Germany_support,min_threshold = 0.01,metric='confidence',num_itemsets=1703)
```
